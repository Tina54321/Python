```python
import matplotlib.pyplot as plt

labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [20, 30, 45, 10]
explode = (0.3, 0, 0, 0)
color = ['yellow', 'green', 'red', 'blue']

figure = plt.figure(figsize=(5,5), dpi=100, facecolor='#FCFAF2') 
axes = figure.add_subplot(1,1,1)
axes.pie(values, 
         labels=labels, 
         explode=explode, 
         autopct='%1.1f%%',
         shadow={'ox': -0.01, 'oy': -0.01},    # o = offset
         startangle=180, 
         colors=color)

plt.show()
```
